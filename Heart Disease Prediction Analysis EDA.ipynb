### Joint plots

this is done to again analysis the multiple columns relations through a grapgh and how they are dynamically chnaging when one changes or the other. in this we get the searate dist plots for the parameters as well and for the relation as well we get and through this we can also conclude that which of the two parameters are in connection or not at all connected

The plot includes a scatterplot of the data, with a regression line showing the trend, as well as histograms or kernel density plots of the individual variables on the x- and y-axes.

```python
sns.jointplot(x=df['Age'], y=df['Age'], data=df) #since are related hence we get a single staright line for the same.
```

```python
sns.jointplot(x=df['Age'], y=df['HeartDisease'], data=df) #signifying that both are not at all connected. 
```

```python
sns.jointplot(x=df['MaxHR'], y=df['Age'], data=df) #to get the defined relationship line that best fits the whole concept 
```

```python
sns.jointplot(x=df['MaxHR'], y=df['Age'], data=df, kind='reg') #to get the defined relationship line that best fits the whole concept 
```

this line is definitely showing the inverse relationship between the two

```python
sns.jointplot(x=df['MaxHR'], y=df['Age'], data=df, kind='hex') #the graph converges at the center which means that at the center the density is ofc more.
```

```python
sns.jointplot(x=df['HeartDisease'], y=df['Cholesterol'], data=df, kind='reg')
```

it has a relationship as we can see that where the cholesterol is more there is the more risk of getting the heart disease and hence this is the whole thing potrayed from this grph

```python
sns.jointplot(x=df['MaxHR'], y=df['Cholesterol'], data=df, kind='reg') 
```
